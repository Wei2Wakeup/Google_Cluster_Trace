## ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
when = pd.date_range(start=pd.datetime(2017,1,1),freq='30Min',periods=len(hfhourly_count['Req Num']))
df =pd.DataFrame(when[526:526+336*2], columns=['Date'])
df['Value'] = hfhourly_count2['Req Num'].astype(np.float64).tolist()
indexed_df = df.set_index('Date')
ts=indexed_df['Value']
plt.figure(figsize=(15,5))
plt.plot(ts)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(window=24,center=False).mean() 
    rolstd = timeseries.rolling(window=24,center=False).std()

    #Plot rolling statistics:
    plt.figure(figsize=(15,5))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
test_stationarity(ts)
```

```python
ts_log = np.log(ts)
test_stationarity(ts_log)
```

```python
ts_log_diff = ts_log - ts_log.shift()
plt.plot(ts_log_diff)
plt.show()
```

```python
ts_log_diff.dropna(inplace=True)
test_stationarity(ts_log_diff)
```

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(ts_log_diff, nlags=20)
lag_pacf = pacf(ts_log_diff, nlags=20, method='ols')


plt.figure(figsize=(15,5))
#Plot ACF: 
plt.subplot(121) 
plt.stem(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

#Plot PACF:

plt.subplot(122)
plt.stem(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
plt.show()
```

```python
model = ARIMA(ts_log, order=(0, 1, 1))  
results_ARIMA = model.fit(disp=-1)  
plt.figure(figsize=(15,5))
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.show()
```

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA_log = pd.Series(ts_log.iloc[0], index=ts_log.index)
```

```python
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.figure(figsize=(15,5))
plt.plot(ts_log)
plt.plot(predictions_ARIMA_log)
plt.xlabel('Date',fontsize=18)
plt.ylabel('Log value of count',fontsize=18)
plt.show()
```

```python
size = int(len(ts_log) - 48)
train, test = ts_log[0:size], ts_log[size:len(ts_log)]
history = [x for x in train]
predictions = list()
```

```python
for t in range(len(test)):
    model = ARIMA(history, order=(0,1,1))
    model_fit = model.fit(disp=-1)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(float(yhat))
    obs = test[t]
    history.append(obs)
    #print('predicted=%f, expected=%f' % (np.exp(yhat), np.exp(obs)))
```

```python
pre_df=pd.DataFrame(test.index.values,columns=['Date'])
pre_df['Value']=predictions
indexed_Pre= pre_df.set_index('Date')
ts_pred=indexed_Pre['Value']
```

```python
ts_pred_org=np.exp(ts_pred)
ts_test_org=ts[size:len(ts_log)]
plt.figure(figsize=(10,5))
plt.plot(ts_test_org,label='Test')
plt.plot(ts_pred_org,label='Prediction')
plt.legend(loc='best')
plt.show()
res=abs(ts_test_org-ts_pred_org)/ts_test_org
print 'MAPE: {}'.format(res.mean())
```

## Last week data

### 1 Direct use last week data

```python
train=hfhourly_count2.iloc[0:336]['Req Num']
test=hfhourly_count2.iloc[336:336*2]['Req Num']
# MAPE
res = abs(np.array(train)-np.array(test))/np.array(test)
res.mean()
```

### 2 Weigted with MA

```python
ma =  hfhourly_count2.iloc[335:336*2]['Req Num'].shift()
pred = np.array(ma.dropna())*0.8 + np.array(train) *0.2
```

```python
# MAPE
res = abs(pred-np.array(test))/np.array(test)
res.mean()
```

### 3 smoothing last week

```python
import pyqt_fit.nonparam_regression as smooth
from pyqt_fit import npr_methods
predict=[]
for i in range(336/48):
    xs=np.array(range(48))
    ys=np.array(hfhourly_count2.iloc[i*48:i*48+48]['Req Num'])
    k1 = smooth.NonParamRegression(xs, ys, method=npr_methods.LocalPolynomialKernel(q=1))
    k1.fit()
    res=k1(xs)
    predict.append(res)
```

```python
pred=np.concatenate(predict)
```

```python
# MAPE
res = abs(pred-np.array(test))/np.array(test)
res.mean()
```

### 5 last week data with ratio between MA

```python
when = pd.date_range(start=pd.datetime(2017,1,1),freq='30Min',periods=len(hfhourly_count['Req Num']))
df =pd.DataFrame(when, columns=['Date'])
df['Value'] = hfhourly_count['Req Num'].astype(np.float64)
indexed_df = df.set_index('Date')
ts=indexed_df['Value']
```

```python
ts_roll=ts.rolling(window=20,center=False).mean().shift()
```

```python
ts_roll2=ts_roll[526+336:526+336*2]
ts_roll1=ts_roll[526:526+336]
```

```python
pred = np.array(ts_roll2)/np.array(ts_roll1)*ts[526:526+336]
# MAPE
res = abs(np.array(pred)-np.array(ts[526+336:526+336*2]))/np.array(ts[526+336:526+336*2])
res.mean()
```

### 6 Smoothed last week data with ratio between MA

```python
import pyqt_fit.nonparam_regression as smooth
from pyqt_fit import npr_methods
predict=[]
for i in range(336/48):
    xs=np.array(range(48))
    ys=np.array(hfhourly_count2.iloc[i*48:i*48+48]['Req Num'])
    k1 = smooth.NonParamRegression(xs, ys, method=npr_methods.LocalPolynomialKernel(q=1))
    k1.fit()
    res=k1(xs)
    predict.append(res)

pred = np.array(ts_roll2)/np.array(ts_roll1)*np.concatenate(predict)
# MAPE
res = abs(np.array(pred)-np.array(ts[526+336:526+336*2]))/np.array(ts[526+336:526+336*2])
res.mean()
```
